```python
import polars as pl
import os
from dotenv import load_dotenv

df = pl.read_excel("raw_data/310101.xlsx", sheet_name="Data1")
aus_population_df = pl.DataFrame({
    "name": df.select(pl.col("__UNNAMED__0").slice(9)),
    "birth": df.select(pl.col("Births ;  Australia ;").slice(9).cast(pl.Float64)),
    "death": df.select(pl.col("Deaths ;  Australia ;").slice(9).cast(pl.Float64)),
    "interstate_arrival": df.select(pl.col("Interstate Arrivals ;  Australia ;").slice(9).cast(pl.Float64)),
    "interstate_departure": df.select(pl.col("Interstate Departures ;  Australia ;").slice(9).cast(pl.Float64)),
    "oversea_arrival": df.select(pl.col("Overseas Arrivals ;  Australia ;").slice(9).cast(pl.Float64)),
    "oversea_departure": df.select(pl.col("Overseas Departures ;  Australia ;").slice(9).cast(pl.Float64)),
    "migration_adjustment": df.select(pl.col("Migration Adjustment ;  Australia ;").slice(9).cast(pl.Float64)),
    "population": df.select(pl.col("Estimated Resident Population (ERP) ;  Australia ;").slice(9).cast(pl.Float64))})
aus_population_df = aus_population_df.with_columns(pl.col("migration_adjustment").fill_null(0))
aus_population_df = aus_population_df.with_columns(
    pl.col(['birth',
 'death',
 'interstate_arrival',
 'interstate_departure',
 'oversea_arrival',
 'oversea_departure',
 'migration_adjustment',
 'population']) * 900
)

load_dotenv()

print(os.getenv("GOOGLE_APPLICATION_CREDENTIALS"))
target_path = "gs://australia-analytics/population/death-birth.parquet"

aus_population_df.write_parquet(target_path, storage_options={"service_account_key_file": os.getenv("GOOGLE_APPLICATION_CREDENTIALS")})
```

# Interstate Migration

```python
ROWS_SKIPPED = 9
states = [{"state": "NSW", "full_state": "New South Wales"},
          {"state": "VIC", "full_state": "Victoria"},
          {"state": "QLD", "full_state": "Queensland"},
          {"state": "SA", "full_state": "South Australia"},
          {"state": "WA", "full_state": "Western Australia"},
          {"state": "TAS", "full_state": "Tasmania"},
          {"state": "NT", "full_state": "Northern Territory"},
          {"state": "ACT", "full_state": "Australian Capital Territory"},
          ]

int_arr_df = pl.read_excel("raw_data/3101016A.xlsx", sheet_name="Data1")
int_dep_df = pl.read_excel("raw_data/3101016B.xlsx", sheet_name="Data1")
int_arr_df
```

```python
# Create an empty dataframe
aus_interstate_df = pl.DataFrame(schema=["date", "state", "interstate_arrival", "interstate_departure"])
# For each state get their migration data
for state in states:
    int_migration = pl.DataFrame({
        "date": int_arr_df.select(pl.col("__UNNAMED__0").slice(9)),
        "state": state["state"],
        "interstate_arrival": int_arr_df.select(pl.col(f"Interstate Arrivals ;  {state["full_state"]} ;").slice(ROWS_SKIPPED).cast(pl.Int32)),
        "interstate_departure": int_dep_df.select(pl.col(f"Interstate Departures ;  {state["full_state"]} ;").slice(ROWS_SKIPPED).cast(pl.Int32)),
    })
    if aus_interstate_df.is_empty():
        aus_interstate_df = int_migration.clone()
    else:
        aus_interstate_df = pl.concat([aus_interstate_df, int_migration])


target_path = "gs://australia-analytics/population/aus_interstate_migration.parquet"

aus_interstate_df.write_parquet(target_path, storage_options={"service_account_key_file": os.getenv("GOOGLE_APPLICATION_CREDENTIALS")},)
print("Successfully saved the immigration to Cloud Storage!")
```

```python
state_files = [{"state": "NSW", "file": "raw_data/3101051.xlsx"},
          {"state": "VIC", "file": "raw_data/3101052.xlsx"},
          {"state": "QLD", "file": "raw_data/3101053.xlsx"},
          {"state": "SA", "file": "raw_data/3101054.xlsx"},
          {"state": "WA", "file": "raw_data/3101055.xlsx"},
          {"state": "TAS", "file": "raw_data/3101056.xlsx"},
          {"state": "NT", "file": "raw_data/3101057.xlsx"},
          {"state": "ACT", "file": "raw_data/3101058.xlsx"},
          ]

male_column_names = [{"col_name": f"Estimated Resident Population ;  Male ;  {age} ;" if age < 100  else f"Estimated Resident Population ;  Male ;  100 and over ;", "age": age, "gender": "male" } for age in range(101)]
female_column_names = [{"col_name": f"Estimated Resident Population ;  Female ;  {age} ;" if age < 100  else f"Estimated Resident Population ;  Female ;  100 and over ;", "age": age, "gender": "female" } for age in range(101)]

column_names = male_column_names + female_column_names

pop_age_df = pl.DataFrame(schema=["date", "age", "gender", "count",])
for state in state_files:
    pop_data_df = pl.read_excel(state["file"], sheet_name="Data1")
    for column in column_names:
        state_pop_df = pl.DataFrame({
            "date": pop_data_df.select(pl.col("__UNNAMED__0").slice(ROWS_SKIPPED)),
            "state": state["state"],
            "gender": column["gender"],
            "age": column["age"],
            "type": f"{column["age"]} yo" if column["age"] < 100 else "100+ yo",
            "count": pop_data_df.select(pl.col(column["col_name"]).slice(ROWS_SKIPPED).cast(pl.Int32)),
        })
        if pop_age_df.is_empty():
            pop_age_df = state_pop_df.clone()
        else:
            pop_age_df = pl.concat([pop_age_df, state_pop_df])

target_path = "gs://australia-analytics/population/aus_gender_population.parquet"

pop_age_df.write_parquet(target_path, storage_options={"service_account_key_file": os.getenv("GOOGLE_APPLICATION_CREDENTIALS")},)
print("Successfully saved the immigration to Cloud Storage!")

```
